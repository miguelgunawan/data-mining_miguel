7. I have a character with this specification:
defense_overall 6.0<br>
offense_overall 7.0<br>
skill_effect_overall 4.0<br>
difficulty_overall 3.0<br>
movement_spd 250<br>
magic_defense 9<br>
mana 380<br>
hp_regen 37<br>
physical_atk 119<br>
physical_defense 24<br>
hp 2689<br>
attack_speed 0.75<br>
mana_regen 16.0<br>

Please predict its role using DecisionTree classification algorithm <br>

```python
# 7

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Encode the 'role' column
label_encoder = LabelEncoder()
df6['role'] = label_encoder.fit_transform(df6['role'])

# Separate features and target variable
x = df6.drop(['role', 'hero_name','win_rate','pick_rate','ban_rate','release_year'], axis=1)
y = df6['role']

# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Create a Decision Tree Classifier
classifier = DecisionTreeClassifier()

# Train the model
classifier.fit(x_train, y_train)
# Make predictions on the test set
y_pred = classifier.predict(x_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Predict the role for the given character
prediction=classifier.predict([[6.0,7.0,4.0,3.0,250,9,380,37,119,24,2689,0.75,16.0]])
# Get the predicted role (dicoding)
predicted_role_name = label_encoder.inverse_transform(prediction)
print("Predicted Role:", predicted_role_name[0])
```
