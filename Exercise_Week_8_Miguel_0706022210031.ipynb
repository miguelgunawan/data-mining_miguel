## Data Preprocessing

For this case study, we only use income and spending score

```python
# create new dataframe that consist only annual income and spending score
df = df[['AnnualIncome', 'SpendingScore']]

print(df)
```

Before we analyze our data, we need to scale the data. You can use MinMaxScaler or StandardScaler.

```python
# scale dataset with standard scaler
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df), columns=["AnnualIncome", "SpendingScore"])

df_scaled.head()
```

```python
# get stastical information on the new dataset
df_scaled.describe()
```

Then, we check the inisial position for our sample. You may use pairplot or scattered chart for the visualization option.

```python
# Visualize the position
plt.figure(figsize=(8, 6))
plt.scatter(df_scaled['AnnualIncome'], df_scaled['SpendingScore'])
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.show()
```

```python
# Visualize the position using the original scale
plt.figure(figsize=(8, 6))
plt.scatter(df['AnnualIncome'], df['SpendingScore'])
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.show()
```

## Clustering

The evaluation for chosing the number of K in KMEANS can be done by several methods such as elbow method.

1. using elbow method
In this case study, the below parameter settings is used:
- init: random
- range: (1,11)
- max_iter: 300
- random state: 42

```python
# set the parameter for evaluating K
from sklearn.cluster import KMeans
sse=[]

for i in range(1,11):
  kmeans = KMeans(n_clusters=i, init='random', max_iter=300, random_state=42)
  kmeans.fit(df_scaled)
  sse.append(kmeans.inertia_)
```

```python
# manually visualize the elbow method using plotlib
plt.plot(range(1,11), sse)
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.show()
```

```python
# using kneelocator to find the elbow/knee
from kneed import KneeLocator

kneedle = KneeLocator(range(1,11), sse, curve="convex", direction="decreasing")
print("The best k for you model is", kneedle.elbow)
```

### K-Means Algorithm

For using KMeans clustering algorithm, you need to define the number of cluster, initialization number, number of iteration, and random state.

In this case study, the below parameter settings is used:
- centroid initialization: random
- number of cluster: 5
- number of iteration: 300
- random state: 42

```python
# assign cluster with K-Means algorithm
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=5, init='random', max_iter=300, random_state=42)
y_pred = kmeans.fit_predict(df_scaled)
y_pred
```

Visualize your result based on income and spending. You also can show the final position of the centroid.

```python
# visualize the result using scater chart
plt.figure(figsize=(10,7))

sns.scatterplot(x="AnnualIncome", y="SpendingScore", hue=y_pred, palette=["blueviolet", "red", "deepskyblue", "orange", "springgreen"], legend=None, data=df_scaled, s=60)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c="black")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.title("Clustering Result with K-Means Method")
plt.show()
```

```python
# Inverse transform the scaled data to get the original values
df = pd.DataFrame(scaler.inverse_transform(df_scaled), columns=["AnnualIncome", "SpendingScore"])
df['Cluster'] = y_pred

df.head()
```

```python
# Group the data by cluster and calculate the mean spending and annual income for each cluster
cluster_summary = df.groupby(y_pred).agg({
    'SpendingScore': ['mean', 'std', 'min', 'max'],
    'AnnualIncome': ['mean', 'std', 'min', 'max']
})

cluster_summary = cluster_summary.rename_axis('cluster')

cluster_summary

# Kesimpulan
# Cluster 0: Kelompok ini kemungkinan memiliki daya beli yang lebih rendah atau minat yang lebih sedikit pada pembelian besar,
# karena pendapatan mereka rendah. Perusahaan mungkin dapat menawarkan produk yang ramah anggaran namun bernilai bagi segmen ini.

# Cluster 1: Pelanggan ini mungkin memiliki kebiasaan belanja tinggi meskipun pendapatan mereka lebih rendah.
# Menargetkan mereka dengan produk premium namun tetap terjangkau bisa menjadi strategi yang efektif.

# Cluster 2: Individu dalam kelompok ini memiliki kemampuan untuk berbelanja (karena pendapatan yang tinggi)
# namun mungkin tidak tertarik pada pembelian yang sering atau dalam jumlah besar.
# Ini mungkin menunjukkan bahwa mereka cenderung menabung dan tidak menggunakan seluruh pendapatan mereka. Perusahaan dapat mencoba pemasaran yang dipersonalisasi untuk menarik kelompok ini dengan produk-produk mewah atau yang bernilai investasi tinggi.

# Cluster 3**: Ini adalah segmen yang berharga dengan daya beli yang kuat dan keinginan tinggi untuk berbelanja,
# bahkan untuk hal-hal yang tidak mereka butuhkan. Pemasaran premium dan penawaran eksklusif dapat menarik kelompok ini,
# karena mereka cenderung menjadi pelanggan bernilai tinggi.

# Cluster 4**: Pelanggan ini mungkin merupakan kelompok dengan pendapatan menengah dan pengeluaran yang seimbang.
# Strategi pemasaran umum dan produk kelas menengah mungkin cocok untuk segmen ini.
```
